```python
import pandas as pd
```

```python
Gwangju = pd.read_csv('광주광역시.csv')
Daegu = pd.read_csv('대구광역시.csv')
Busan = pd.read_csv('부산광역시.csv')
```

```python
Gwangju.rename(columns={'시도명':'시도'},inplace=True)
```

```python
Gwangju.head(1)
```

```python
Daegu.head(1)
```

```python
Busan.head(1)
```

```python
Gwangju.columns
```

```python
Gwangju.rename(columns={'주차장유형':'유형'},inplace=True)
```

```python
Daegu.rename(columns={'주차장유형(노상_노외_부설)':'유형'},inplace=True)
```

```python
Gwangju.rename(columns={'소재지도로명주소':'주소'},inplace=True)
```

```python
Daegu.rename(columns={'전화번호(주차장관리자)':'전화번호'},inplace=True)
```

```python
Busan.rename(columns={'기관명':'관리기관'},inplace=True)
```

```python
Busan.rename(columns={'365일 상시 무료개방 여부':'종일개방여부'},inplace=True)
```

```python
Gwangju.head(1)
```

```python
Daegu.head(1)
```

```python
Busan.head(1)
```

```python
result = pd.concat([Gwangju,Daegu, Busan])
```

```python
result.head(1)
```

```python
result.columns
```

```python
parking = result.drop(columns=['위치 정보', '01월29일(토)', '01월30일(일)', '01월31일(월)','02월01일(화)',
                               '02월02일(수)','위도','경도','개방시작시간','개방종료시간'])
```

```python
parking
```

```python
parking['종일개방여부']=parking['종일개방여부'].replace(['Y','X'],['종일개방','일부개방'])
```

```python
parking
```

```python
parking['구군']=parking['구군'].fillna('서구')
```

```python
parking
```

```python
parking = parking.set_index(['시도','유형','구군'])
```

```python
parking
```

```python
parking.to_pickle('광역시 무료 주차장.pickle')
```
